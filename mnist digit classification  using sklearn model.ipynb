```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train_x, y_train, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, Y_train)
y_pred = classifier.predict(X_test) # predict our file test data
```

```python
print("Model accuracy is: {0:.3f}%".format(accuracy_score(Y_test, y_pred) * 100))
```
